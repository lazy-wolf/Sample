```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardize the features (data normalization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Function to perform KMeans clustering with different parameters
def perform_kmeans(n_clusters_val, init_val, n_init_val, max_iter_val):
    kmeans = KMeans(n_clusters=n_clusters_val, init=init_val, n_init=n_init_val, max_iter=max_iter_val, random_state=42)
    kmeans.fit(X_scaled)
    labels = kmeans.labels_
    return labels

# Experiment with different parameters
n_clusters_values = [2, 3, 4, 5]
init_values = ['k-means++', 'random']
n_init_values = [10, 20, 30]
max_iter_values = [100, 200, 300]

for n_clusters in n_clusters_values:
    for init_method in init_values:
        for n_init in n_init_values:
            for max_iter in max_iter_values:
                labels = perform_kmeans(n_clusters, init_method, n_init, max_iter)
                # Visualize or analyze the results here (e.g., using evaluation metrics)

# Example visualization (for demonstration purposes)
# Replace this with actual analysis and visualizations
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, random_state=42)
kmeans.fit(X_scaled)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.datasets import load_wine
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Load the Wine dataset
wine = load_wine()
X = wine.data
y = wine.target

# Standardize the features (data normalization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Function to perform Agglomerative Clustering with different parameters
def perform_agglomerative(n_clusters_val, affinity_val, linkage_val):
    agg_clustering = AgglomerativeClustering(n_clusters=n_clusters_val, affinity=affinity_val,
                                             linkage=linkage_val)
    labels = agg_clustering.fit_predict(X_scaled)
    return labels

# Experiment with different parameters
n_clusters_values = [3, 4, 5, 6]
affinity_values = ['euclidean']
linkage_values = ['ward', 'complete', 'average', 'single']

for n_clusters in n_clusters_values:
    for affinity_method in affinity_values:
        for linkage_method in linkage_values:
            labels = perform_agglomerative(n_clusters, affinity_method, linkage_method)
            # Analyze the results here using evaluation metrics or visualizations

# Example visualization (for demonstration purposes)
# Replace this with actual analysis and visualizations
agg_clustering = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
labels = agg_clustering.fit_predict(X_scaled)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('Agglomerative Clustering on Wine Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
# Load the Wine dataset
wine = load_wine()
X = wine.data

# Standardize the features (data normalization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Function to evaluate clustering algorithms using Silhouette Score and Davies-Bouldin Index
def evaluate_clustering(labels, data):
    silhouette = silhouette_score(data, labels)
    davies_bouldin = davies_bouldin_score(data, labels)
    return silhouette, davies_bouldin

# Perform KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
kmeans_silhouette, kmeans_davies_bouldin = evaluate_clustering(kmeans_labels, X_scaled)

# Perform Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
agg_labels = agg_clustering.fit_predict(X_scaled)
agg_silhouette, agg_davies_bouldin = evaluate_clustering(agg_labels, X_scaled)

# Compare results
print("KMeans:")
print(f"Silhouette Score: {kmeans_silhouette}")
print(f"Davies-Bouldin Index: {kmeans_davies_bouldin}")
print("\nAgglomerative Clustering:")
print(f"Silhouette Score: {agg_silhouette}")
print(f"Davies-Bouldin Index: {agg_davies_bouldin}")
```
